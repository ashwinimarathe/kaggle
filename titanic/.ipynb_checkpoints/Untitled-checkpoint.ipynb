```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
train = pd.read_csv('train.csv', index_col=0)
train.head()
```

```python
sns.heatmap(train[['Survived','Age','Fare','Parch','SibSp']].corr(), annot=True)
```

```python
def compare_variables(variable1, variable2):
    print(train[[variable1,variable2]][train[variable1].isnull()==False].groupby([variable1],
        as_index=False).mean().sort_values(by=variable2, ascending=False))
    sns.FacetGrid(train, col=variable2).map(sns.distplot, variable1)

```

```python
compare_variables('Age','Survived')
```

```python
compare_variables('Pclass','Survived')
```

```python
compare_variables('Parch','Survived')
```

```python
compare_variables('SibSp','Survived')
```
